## gemma2:2b

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = (
    OllamaEmbeddings(model = "gemma2:2b") # by default it uses llama2 model
)

embeddings
```

```python
res = embeddings.embed_documents(
    [
        "I recently completed my Post Graduation Diploma in Artificial Intelligence from CDAC pune.", 
        "I have completed my graduation from Dy Patil RAIT, Navi Mumbai in year 2023."
    ]
)

res
```

```python
embeddings.embed_query("I recently completed my Post Graduation Diploma in Artificial Intelligence from where.")
```
