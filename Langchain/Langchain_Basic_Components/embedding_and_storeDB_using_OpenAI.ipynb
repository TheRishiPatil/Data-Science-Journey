# Vector Embedding

```python
import os
from dotenv import load_dotenv

load_dotenv() # load all the environment variables
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model= "text-embedding-3-large")
embeddings
```

```python
text = "I'm Rishikesh Krishna Patil, from New Mumbai."

embedded_text = embeddings.embed_query(text)
embedded_text
```

```python
len(embedded_text)
```

## set custom Dimensions

```python
embeddings = OpenAIEmbeddings(model= "text-embedding-3-large", dimensions = 1080)
embedded_text = embeddings.embed_query(text)
embedded_text
```

```python
len(embedded_text)
```

# Vector StoreDB

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader('data/intro.txt')
text_document = loader.load()
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(text_document, embeddings)
db
```

```python
query = "I'm Rishikesh Krishna Patil, from New Mumbai."
res = db.similarity_search(query)
res
```
