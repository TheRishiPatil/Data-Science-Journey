```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder as ohe
```

```python
df = pd.DataFrame(['red', 'green', 'blue', 'yellow', 'red', 'yellow', 'green'], columns = ['colors'])
df
```

```python
encoder = ohe()

encoded = encoder.fit_transform(df[['colors']]).toarray()

df[encoder.get_feature_names_out()] = encoded
df
```

```python
import seaborn as sns
df2 = sns.load_dataset('tips')
df2
```

```python
df2[encoder.get_feature_names_out()] = encoder.fit_transform(df2[['sex', 'smoker']]).toarray()
df2
```
