```python
from sklearn.datasets import make_classification
# make_classification --> used to create dataset

from imblearn.over_sampling import SMOTE
# SMOTE --> it is responsible for connecting nearest points and creating new datapoints

import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples = 1000, n_redundant = 0, n_features = 2, n_clusters_per_class = 1, weights = [0.90], random_state = 12)

# n_samples --> total datapoints
# n_features --> feature columns
# n_clusters_per_class --> cluster (group) per class
# weights --> one class weight compared to other
```

```python
df1 = pd.DataFrame(X, columns = ['f1', 'f2'])
df2 = pd.DataFrame(y, columns = ['target'])

final_df = pd.concat([df1, df2], axis = 1)

plt.scatter(final_df['f1'], final_df['f2'], c = final_df['target'])
final_df.target.value_counts()
```

```python
overSampling = SMOTE()
df1, df2 = overSampling.fit_resample(final_df[['f1', 'f2']], final_df['target'])
# fit_resample() --> performs two essential tasks: fitting the SMOTE model to the data and then applying the resampling technique to balance the classes.

final_df = pd.concat([df1, df2], axis = 1)

plt.scatter(final_df['f1'], final_df['f2'], c = final_df['target'])
final_df.target.value_counts()
```
