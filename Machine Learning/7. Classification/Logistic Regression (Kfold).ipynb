```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df['target'] = dataset.target
df
```

```python
new_df = df[df['target'] != 2]
new_df
```

```python
X = new_df.drop('target', axis = 1)
y = new_df.target
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(max_iter = 200)
classification.fit(X_train, y_train)
y_test_prediction = classification.predict(X_test)
```

## Confusion Metrics, Accuracy, Classification Report

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(confusion_matrix(y_test, y_test_prediction))
print(accuracy_score(y_test, y_test_prediction))
print(classification_report(y_test, y_test_prediction))
```

## Cross Validation

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits = 5, shuffle = True)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classification, X_train, y_train, scoring = 'accuracy', cv = cv)
np.mean(scores)
```
